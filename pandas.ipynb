```python
tst = pd.read_csv(tst_file, index_col=0)  #시험 데이터 로드
print(tst.shape)
tst.head()
```

```python
df = pd.concat([trn, tst], axis=0) #concat xix = 0 일때 ==> 위아래로 합치기
print(df.shape)
df.tail()
```

```python
df.fillna(-1, inplace=True) #결측값을 -1로 바꾼다. inplace = True를 설정함으로써
#또 다른 객체를 반환하지 않고 기존 객체를 수정한다. 즉, 기존 데이터 프레임을 보존하지
#않고 바로 결측값을 해당값으로 수정된다.
df.tail()
```

```python
df['nObserve'].hist(bins=30) #히스토그램 출력
```

```python
df['nObserve'] = df['nObserve'].apply(np.log1p) #log1p = log(1+z) => -무한대 예방
```

```python
df['nObserve'].hist(bins=30)
```

```python
df['d_dered_u'] = df['dered_u'] - df['u']  #추가하기
df['d_dered_g'] = df['dered_g'] - df['g']  #추가하기
df['d_dered_r'] = df['dered_r'] - df['r']  #추가하기
df['d_dered_i'] = df['dered_i'] - df['i']  #추가하기
df['d_dered_z'] = df['dered_z'] - df['z']  #추가하기
df['d_dered_rg'] = df['dered_r'] - df['dered_g']  #추가하기
df['d_dered_ig'] = df['dered_i'] - df['dered_g']  #추가하기
df['d_dered_zg'] = df['dered_z'] - df['dered_g']  #추가하기
df['d_dered_ri'] = df['dered_r'] - df['dered_i']  #추가하기
df['d_dered_rz'] = df['dered_r'] - df['dered_z']  #추가하기
df['d_dered_iz'] = df['dered_i'] - df['dered_z']  #추가하기
df['d_obs_det'] = df['nObserve'] - df['nDetect']  #추가하기
print(df.shape)
df.head()
```

```python
df.corr().style.background_gradient() #각 컬럼마다 배경색을 수치에 비례하여 입힌다.
```

```python
df.drop(['airmass_z', 'airmass_i', 'airmass_r', 'airmass_g', 'u', 'g', 'r', 'i', 'nDetect', 'd_dered_rg', 'd_dered_ri'], 
        axis=1, inplace=True) #삭제
print(df.shape)
df.head()
```

```python
df.to_csv(feature_file) # "df"내용 다음 경로로 저장
# feature_file = data_dir / 'feature.csv'
```
