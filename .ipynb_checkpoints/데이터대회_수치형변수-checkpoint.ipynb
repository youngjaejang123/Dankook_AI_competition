```python
from matplotlib import rcParams, pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
import seaborn as sns
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from warnings import simplefilter
```

```python
rcParams['figure.figsize'] = (16, 8)
plt.style.use('fivethirtyeight')
simplefilter('ignore')
```

```python
#대회 데이터 로드
data_dir = Path('../input/')
trn_file = data_dir / 'train.csv'
seed = 42
```

```python
trn = pd.read_csv(trn_file, index_col=0)
print(trn.shape)
trn.head()
```

```python
trn.info()
#독립변수를 포함하여 전부(총18개)가 수치형변수이다.
#독립변수 : 연구자가 의도적으로 변화시키는 변수 >> 즉 다른 변수와 영향이 없는 변수
#종속변수 : 연구자가 독립변수의 변화에 따라 어떻게 변하는지 알고 싶은 변수 >> 영향존재
#수치형변수 : 정수/실수
#범주형변수 : 성별, 이름 등등
#이진변수 : 참/거짓 ex) 암의 여부
#또한 null 값이 없다.
```

```python
trn.describe()
```

```python
trn.hist(bins=50) #각 데이터 컬럼의 분포형 확인
```

```python
trn.fillna(0, inplace=True)
trn.info()
```

```python
target_col = 'u'
pred_col = 'pred'
num_cols = [x for x in trn.columns if trn[x].dtype in [np.int64, np.float64] and x != target_col]
print(num_cols)
```

```python
trn['u'].hist(bins=50)
```

```python
sns.pairplot(data=trn, vars=['dered_u', 'airmass_u'] + [target_col], size = 5)
```

```python
trn.corr().style.background_gradient() #상관관계 표현 1에 가까울수록 완전일치
#0에 가까울수록 아예 다름, -1에 가까울수록 반대로 완전히 일치
```

```python
#선형회귀 모델 학습
#RMSE/RMLSE 손실 함수 정의  // 손실함수 = 오차를 의미
rmse = lambda y, p: np.sqrt(mean_squared_error(y, p))
rmlse = lambda y, p: np.sqrt(mean_squared_error(np.log1p(y), np.log1p(p)))
#y는 실제값, p는 예측값
```

```python
#변수 가공없이 선형회귀 모델학습
lr = LinearRegression()
lr.fit(trn[num_cols], trn[target_col])
trn[pred_col] = lr.predict(trn[num_cols])

print(f' RMSE:\t{rmse(trn[target_col], trn[pred_col]):12.2f}')
```

```python
sns.pairplot(data=trn, vars=[pred_col, target_col], size=4, plot_kws={'alpha': .5})
#class의 실제값과 예측값
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(trn[target_col], trn[pred_col], alpha=.5)
plt.xlabel('u')
plt.ylabel('prediction')
plt.yscale('log')
plt.xscale('log')
```
